### Import Libraries

```python
import os
import time
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, f1_score
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.layers import Input, LSTM, Dense, Dropout , RepeatVector
import joblib
import matplotlib.pyplot as plt
import seaborn as sns
import mlflow
import mlflow.keras
import mlflow.sklearn
```

## AutoEncoder

```python
# Define gestures
GESTURES = ["swipe", "push-pull", "circular", "unidentified"]
data_seq = '3'
data_path = f'../data/processed/{data_seq}/gesture_data.npz'

# Load preprocessed data
data = np.load(data_path)
X = data['data']
y = data['labels']

# Normalize the data
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X.reshape(-1, X.shape[-1])).reshape(X.shape)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Autoencoder Definition
def create_autoencoder(input_dim):
    input_layer = Input(shape=(input_dim,))

    # Encoder
    encoded = Dense(128, activation='relu')(input_layer)
    encoded = Dense(64, activation='relu')(encoded)
    encoded = Dense(32, activation='relu')(encoded)

    # Decoder
    decoded = Dense(64, activation='relu')(encoded)
    decoded = Dense(128, activation='relu')(decoded)
    decoded = Dense(input_dim, activation='sigmoid')(decoded)

    # Models
    autoencoder = Model(inputs=input_layer, outputs=decoded)
    encoder = Model(inputs=input_layer, outputs=encoded)

    autoencoder.compile(optimizer='adam', loss='mse')
    return autoencoder, encoder

input_dim = X_train.shape[1]  # Number of features

# MLflow Experiment Tracking
mlflow.set_experiment("Gesture Recognition Experiment")

with mlflow.start_run(run_name="Autoencoder + Classifier"):
    # Create Autoencoder and Encoder
    autoencoder, encoder = create_autoencoder(input_dim)

    # Train Autoencoder
    print("Training Autoencoder...")
    autoencoder.fit(
        X_train, X_train,
        epochs=50,
        batch_size=32,
        validation_data=(X_test, X_test),
        verbose=1
    )
    
    
    # Log Autoencoder Model with Input Example
    #input_example_autoencoder = X_test[:1]  # Example batch from test set
    mlflow.keras.log_model(autoencoder, "autoencoder_model"
                           #, input_example=input_example_autoencoder
                           )


    # Save Encoder and Scaler
    model_dir = f"../models/{data_seq}/"
    os.makedirs(model_dir, exist_ok=True)
    encoder.save(os.path.join(model_dir, "encoder_model.h5"))
    joblib.dump(scaler, os.path.join(model_dir, "scaler.pkl"))

    # Extract Latent Features
    print("Extracting Latent Features...")
    X_train_latent = encoder.predict(X_train)
    X_test_latent = encoder.predict(X_test)

    # Define Classifier Model
    def create_classifier(input_dim, num_classes):
        model = Sequential([
            Dense(64, activation='relu', input_shape=(input_dim,)),
            Dropout(0.2),
            Dense(32, activation='relu'),
            Dense(num_classes, activation='softmax')
        ])
        model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
        return model

    num_classes = len(np.unique(y))
    classifier = create_classifier(X_train_latent.shape[1], num_classes)

    # Train Classifier
    print("Training Classifier...")
    classifier.fit(
        X_train_latent, y_train,
        epochs=100,
        batch_size=32,
        validation_data=(X_test_latent, y_test),
        verbose=1
    )

    ## Log Classifier Model with Input Example
    #input_example_classifier = X_test_latent[:1]
    mlflow.keras.log_model(classifier, "classifier_model", 
                           #input_example=input_example_classifier
                           )

    # Evaluation
    def evaluate_model(model, X_test, y_test):
        y_pred = np.argmax(model.predict(X_test), axis=1)

        # Classification Report
        report = classification_report(y_test, y_pred, target_names=GESTURES[:-1], output_dict=True)
        print("\nEvaluation Results:")
        print(classification_report(y_test, y_pred, target_names=GESTURES[:-1]))

        # Accuracy
        accuracy = accuracy_score(y_test, y_pred)
        print(f"Accuracy: {accuracy * 100:.2f}%")

        # Confusion Matrix
        conf_matrix = confusion_matrix(y_test, y_pred)
        plt.figure(figsize=(8, 6))
        sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=GESTURES[:-1], yticklabels=GESTURES[:-1])
        plt.title("Confusion Matrix")
        plt.xlabel("Predicted")
        plt.ylabel("True")
        plt.show()

        # Log Metrics
        mlflow.log_metric("accuracy", accuracy)
        for label, metrics in report.items():
            if isinstance(metrics, dict):
                for metric_name, value in metrics.items():
                    mlflow.log_metric(f"{label}_{metric_name}", value)

    evaluate_model(classifier, X_test_latent, y_test)

    # Save Classifier
    classifier.save(os.path.join(model_dir, "classifier_model.h5"))
    print("Autoencoder and Classifier Training Complete.")

    # Example Inference
    def predict_gesture(rssi_sequence):
        rssi_sequence = scaler.transform(rssi_sequence.reshape(1, -1))
        latent_features = encoder.predict(rssi_sequence)
        prediction = classifier.predict(latent_features)
        return np.argmax(prediction)

    sample_rssi = X_test[0]
    predicted_gesture = predict_gesture(sample_rssi)
    print(f"\nPredicted Gesture: {GESTURES[predicted_gesture]}, True Label: {GESTURES[y_test[0]]}")
```

## AutoEncoder + LSTM

```python

# Define gestures
GESTURES = ["swipe", "push-pull", "circular", "unidentified"]
data_seq = '3'
data_path = f'../data/processed/{data_seq}/gesture_data.npz'

# Load preprocessed data
data = np.load(data_path)
X = data['data']
y = data['labels']

# Normalize the data
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X.reshape(-1, X.shape[-1])).reshape(X.shape)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Autoencoder Definition
def create_autoencoder(input_dim):
    input_layer = Input(shape=(input_dim,))

    # Encoder
    encoded = Dense(128, activation='relu')(input_layer)
    encoded = Dense(64, activation='relu')(encoded)
    encoded = Dense(32, activation='relu')(encoded)

    # Decoder
    decoded = Dense(64, activation='relu')(encoded)
    decoded = Dense(128, activation='relu')(decoded)
    decoded = Dense(input_dim, activation='sigmoid')(decoded)

    # Models
    autoencoder = Model(inputs=input_layer, outputs=decoded)
    encoder = Model(inputs=input_layer, outputs=encoded)

    autoencoder.compile(optimizer='adam', loss='mse')
    return autoencoder, encoder

def create_lstm_classifier(input_dim, num_classes):
    model = Sequential([
        LSTM(64, activation='relu', input_shape=(1, input_dim)),
        Dropout(0.2),
        Dense(32, activation='relu'),
        Dense(num_classes, activation='softmax')
    ])
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model

input_dim = X_train.shape[1]  # Number of features

# MLflow Experiment Tracking
mlflow.set_experiment("Gesture Recognition Experiment")

with mlflow.start_run(run_name="Autoencoder + LSTM"):
    # Create Autoencoder and Encoder
    autoencoder, encoder = create_autoencoder(input_dim)

    # Train Autoencoder
    print("Training Autoencoder...")
    autoencoder.fit(
        X_train, X_train,
        epochs=100,
        batch_size=32,
        validation_data=(X_test, X_test),
        verbose=1
    )

    # Log Autoencoder Model
    #input_example_autoencoder = X_test[:1]
    mlflow.keras.log_model(autoencoder, "autoencoder_with_lstm_model"
                           #, input_example=input_example_autoencoder
                           )

    # Save Encoder and Scaler
    model_dir = f"../models/{data_seq}/"
    os.makedirs(model_dir, exist_ok=True)
    encoder.save(os.path.join(model_dir, "encoder_model_lstm.h5"))
    joblib.dump(scaler, os.path.join(model_dir, "scaler_lstm.pkl"))

    # Extract Latent Features
    print("Extracting Latent Features...")
    X_train_latent = encoder.predict(X_train)
    X_test_latent = encoder.predict(X_test)

    # Reshape for LSTM
    X_train_latent = X_train_latent.reshape(-1, 1, X_train_latent.shape[-1])
    X_test_latent = X_test_latent.reshape(-1, 1, X_test_latent.shape[-1])

    # Define LSTM Classifier
    num_classes = len(np.unique(y))
    lstm_classifier = create_lstm_classifier(X_train_latent.shape[-1], num_classes)

    # Train LSTM Classifier
    print("Training LSTM Classifier...")
    lstm_classifier.fit(
        X_train_latent, y_train,
        epochs=100,
        batch_size=32,
        validation_data=(X_test_latent, y_test),
        verbose=1
    )

    # Log LSTM Classifier
    #input_example_classifier = X_test_latent[:1]
    mlflow.keras.log_model(lstm_classifier, "lstm_classifier_model"
                           #, input_example=input_example_classifier
                           )

    # Evaluation
    def evaluate_model(model, X_test, y_test):
        y_pred = np.argmax(model.predict(X_test), axis=1)

        # Classification Report
        report = classification_report(y_test, y_pred, target_names=GESTURES[:-1], output_dict=True)
        print("\nEvaluation Results:")
        print(classification_report(y_test, y_pred, target_names=GESTURES[:-1]))

        # Accuracy
        accuracy = accuracy_score(y_test, y_pred)
        print(f"Accuracy: {accuracy * 100:.2f}%")

        # Confusion Matrix
        conf_matrix = confusion_matrix(y_test, y_pred)
        plt.figure(figsize=(8, 6))
        sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=GESTURES[:-1], yticklabels=GESTURES[:-1])
        plt.title("Confusion Matrix")
        plt.xlabel("Predicted")
        plt.ylabel("True")
        plt.show()

        # Log Metrics
        mlflow.log_metric("accuracy", accuracy)
        for label, metrics in report.items():
            if isinstance(metrics, dict):
                for metric_name, value in metrics.items():
                    mlflow.log_metric(f"{label}_{metric_name}", value)

    evaluate_model(lstm_classifier, X_test_latent, y_test)

    # Save LSTM Classifier
    lstm_classifier.save(os.path.join(model_dir, "lstm_classifier_model.h5"))
    print("Autoencoder + LSTM Training Complete.")
```

## LSTM based Autoencoder

```python

# Define gestures
GESTURES = ["swipe", "push-pull", "circular", "unidentified"]
data_seq = '3'
data_path = f'../data/processed/{data_seq}/gesture_data.npz'

# Load preprocessed data
data = np.load(data_path)
X = data['data']
y = data['labels']

# Normalize the data
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X.reshape(-1, X.shape[-1])).reshape(X.shape)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# LSTM-based Autoencoder Definition
def create_lstm_autoencoder(input_dim, time_steps):
    # Encoder
    encoder_inputs = Input(shape=(time_steps, input_dim))
    encoded = LSTM(64, activation='relu', return_sequences=True)(encoder_inputs)
    encoded = LSTM(32, activation='relu', return_sequences=False)(encoded)

    # Decoder
    decoded = RepeatVector(time_steps)(encoded)
    decoded = LSTM(32, activation='relu', return_sequences=True)(decoded)
    decoded = LSTM(64, activation='relu', return_sequences=True)(decoded)
    decoded = Dense(input_dim, activation='sigmoid')(decoded)

    # Autoencoder Model
    autoencoder = Model(inputs=encoder_inputs, outputs=decoded)
    encoder = Model(inputs=encoder_inputs, outputs=encoded)

    autoencoder.compile(optimizer='adam', loss='mse')
    return autoencoder, encoder

# Input dimensions for LSTM Autoencoder
time_steps = X_train.shape[1]
input_dim = X_train.shape[2] if len(X_train.shape) > 2 else 1  # Ensure compatibility

# Reshape input data for LSTM
X_train_lstm = X_train.reshape(X_train.shape[0], time_steps, input_dim)
X_test_lstm = X_test.reshape(X_test.shape[0], time_steps, input_dim)

# MLflow Experiment Tracking
mlflow.set_experiment("Gesture Recognition Experiment")

with mlflow.start_run(run_name="LSTM-based Autoencoder"):
    # Create LSTM-based Autoencoder
    autoencoder, encoder = create_lstm_autoencoder(input_dim, time_steps)

    # Train Autoencoder
    print("Training LSTM-based Autoencoder...")
    autoencoder.fit(
        X_train_lstm, X_train_lstm,
        epochs=50,
        batch_size=32,
        validation_data=(X_test_lstm, X_test_lstm),
        verbose=1
    )

    # Log Autoencoder Model
    mlflow.keras.log_model(autoencoder, "lstm_autoencoder_model")

    # Save Encoder and Scaler
    model_dir = f"../models/{data_seq}/"
    os.makedirs(model_dir, exist_ok=True)
    encoder.save(os.path.join(model_dir, "lstm_encoder_model.h5"))
    joblib.dump(scaler, os.path.join(model_dir, "scaler_lstm.pkl"))

    # Extract Latent Features
    print("Extracting Latent Features...")
    X_train_latent = encoder.predict(X_train_lstm)
    X_test_latent = encoder.predict(X_test_lstm)

    # Define Classifier Model
    def create_dense_classifier(input_dim, num_classes):
        model = Sequential([
            Dense(64, activation='relu', input_shape=(input_dim,)),
            Dropout(0.2),
            Dense(32, activation='relu'),
            Dense(num_classes, activation='softmax')
        ])
        model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
        return model

    num_classes = len(np.unique(y))
    classifier = create_dense_classifier(X_train_latent.shape[-1], num_classes)

    # Train Classifier
    print("Training Classifier...")
    classifier.fit(
        X_train_latent, y_train,
        epochs=50,
        batch_size=32,
        validation_data=(X_test_latent, y_test),
        verbose=1
    )

    # Log Classifier Model
    mlflow.keras.log_model(classifier, "dense_classifier_model")

    # Evaluation
    def evaluate_model(model, X_test, y_test):
        y_pred = np.argmax(model.predict(X_test), axis=1)

        # Classification Report
        report = classification_report(y_test, y_pred, target_names=GESTURES[:-1], output_dict=True)
        print("\nEvaluation Results:")
        print(classification_report(y_test, y_pred, target_names=GESTURES[:-1]))

        # Accuracy
        accuracy = accuracy_score(y_test, y_pred)
        print(f"Accuracy: {accuracy * 100:.2f}%")

        # Confusion Matrix
        conf_matrix = confusion_matrix(y_test, y_pred)
        plt.figure(figsize=(8, 6))
        sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=GESTURES[:-1], yticklabels=GESTURES[:-1])
        plt.title("Confusion Matrix")
        plt.xlabel("Predicted")
        plt.ylabel("True")
        plt.show()

        # Log Metrics
        mlflow.log_metric("accuracy", accuracy)
        for label, metrics in report.items():
            if isinstance(metrics, dict):
                for metric_name, value in metrics.items():
                    mlflow.log_metric(f"{label}_{metric_name}", value)

    evaluate_model(classifier, X_test_latent, y_test)

    # Save Classifier
    classifier.save(os.path.join(model_dir, "lstm_dense_classifier_model.h5"))
    print("LSTM-based Autoencoder and Classifier Training Complete.")
```
